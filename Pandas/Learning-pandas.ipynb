Series = A pandas 1D labelled array that can hols any data.(Think of it as an single column in a spreadsheet[1 dimensional])

```python
# We will use a list for a series here:
data = [100, 101, 102]
# Convert this list into series
series = pd.Series(data, index=["a","b","c"]) # Series is a contructor
series
```

```python
# To return a value from the label, we use the loc[] property:
print(series.loc["a"])

# To update a value using label we can use the loc[]:
series.loc["b"] = 98
print(series)
# So, the loc property essentially means "locating a value by label"
```

```python
# To locate by integer postion, we can use "iloc[]":
print(series.iloc[0])

# we need to get index starting from 0.
series.iloc[1] = 101
print(series)
```

```python
# To filter by value, we can use the subscript and then put some condition in it.
data2 = [100, 101, 102, 201, 202]
series2 = pd.Series(data2, index=["a","b","c","d","e"])
print(series2[series2 <= 200])
```

```python
# We will now use a dictionary for a series.
calories = {"day 1" : 1750, "day 2" : 2100, "day 3" : 1700}
# The major difference is that the key of the dictionary will become the label
series3 = pd.Series(calories)
print(series3)
```

```python
#Here, we can use the loc property just like before. 
print(series3.loc["day 1"])

series3.loc["day 3"] += 175
print(series3.loc["day 3"])

print(series3[series3 >= 2000])
```

```python
pokemons = ['bulbasaur' , 'Ivysaur' , 'Venasaur' , 'Charmander' , 'Charmeleon' , 'Charizard' ]
pd_series = pd.Series(pokemons , index=[1,2,3,4,5,6])
print(pd_series)
```

Let's shift towards "DataFrames"  
DataFrames are a tabular data structure with 'Rows' AND 'Columns' (2 Dimensionsal). It is similar to an Excel Spreadsheet.

```python
# We can create a DataFrame by using a Dictionary
data3 = {"Name" : ["Bob" , "Merlin", "Stacy", "Audrey", "Klein"],
         "Age" : [29, 25, 21, 20, 21]
        }
#Now we will convert the dictionary into a DataFrame by using pd.DataFrame.
df = pd.DataFrame(data3, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5"])

df
```

```python
# We can use the loc property here as well.It still means locate by label.
# We can also use 'iloc[]' property as well.
print(df.loc["Employee 2"])
print(df.iloc[0])
```

```python
# To add a new column, we do it like this:
df["Job"] = ["Engineer", "Magician", "Receptionist", "Supervisor", "Student"]

df
```

```python
# To add a new row, we can create a new DataFrame and then concatenate it.
# First, we will create a DataFrame which would be a list of dictionary, in which the 'key' would be the 'column_name' and the 'value' would be the new row value we want to add. We can also add the index for it as a label.

new_row = pd.DataFrame([{"Name": "Alger","Age" : 26 ,"Job": "Navy"}], index= ["Employee 6"])

# To add multiple rows, we can have different dictionaries in the list with the same format of "key","value" pairs seperated by commas, and for label, we can just add the new label seperated by comma in the list.

# new_row = pd.DataFrame([{"Name": "Alger","Age" : 26 ,"Job": "Navy"},
#                         {"Name: "Albert","Age" : 55 ,"Job": "Pilot"},
#                         {"Name: "Andy","Age" : 43 ,"Job": "Artist"}],
#                         index= ["Employee 6", "Employee 7", "Employee 8"])

# Now we will concatenate the two dataframes using 'pd.concat()':
df = pd.concat([df, new_row])

df
```

Now let's see the different selection techniques in pandas.

```python
# Selection By Column
print(df1["Name"].to_string())
```

```python
# To select multiple columns, we can just pass a list:
print(df1[["Name", "Height", "Weight"]].to_string())
```

```python
# Selection by row can be done using the 'loc[]' method like: 
print(df1.loc[0])
print(df1.loc[1])
```

```python
# It is quite hard to remember the default label of every pokemon that pandas has given, so we can just change the label into the name of the pokemon which is easy to look for. We can do that by using the 'index_col' and then entering the column we wnat to use as the index.

df3 = pd.read_csv("pokemon.csv" , index_col="Name")
df3
```

```python
# Now we can easily select a pokemon by the label i.e. Name
df3.loc["Pikachu"]
df3.loc["Charizard"]
```

```python
# If we don't want all the data from the row, we can enter a tuple as an argument of all the columns that we want the data from.
print(df3.loc["Charizard", ("Height","Weight", "Legendary")])
```

```python
# We can also use slice operator (:) to get all the data between two values. For ex - 
print(df3.loc["Charizard" : "Pikachu", ("Height", "Weight")])
```

```python
# To get the first 10 rows with a step of two, we can use the 'iloc[]' property:
print(df3.iloc[0:11:2]) # The second number is excusive.

# we can also use other argument to select the amount of columns we need to show:
print(df3.iloc[0:11:2, 0:3])
```

Now, let us proceed to Filtering

Filtering means keeping the rows that match a certain condition.

```python
dataf = pd.read_csv("pokemon.csv")

#Lets say we want to filter the dataframe by the some height of the pokemons, then we can just do this:
tall_pokemon = dataf[dataf["Height"] >= 2]

# If we want to check for two conditions, we can use OR/AND operators on different conditions in paratheses.
water_pokemon = dataf[(dataf["Type1"] == "Water") | (dataf["Type2"] == "Water")]
ff_pokemon = dataf[(dataf["Type1"] == "Fire") & (dataf["Type2"] == "Flying")]
print(tall_pokemon)
print(" ")
print(water_pokemon)
print(" ")
print(ff_pokemon)
```

Now let us go through aggregate functions.  
***aggregate functions*** - Reduces a set of values into a single summary value. Used to summarize and analyze data.  
Often used by __groupby()__ function.

```python
# For finding the mean of data, we can use the mean() aggregate function. In order to only find use the numeric value for mean, we give an argument of 'numeric_only = True'.

print(dataf.mean(numeric_only=True))
```

```python
# using sum() aggregate fumction
print(dataf.sum(numeric_only=True))
```

```python
# using min() aggregate function
print(dataf.min(numeric_only=True))
```

```python
# using max() aggregate function
print(dataf.max(numeric_only=True))
```

```python
# using count() aggregate function
print(dataf.count())
```

Now comes the aggregate functions that can be applied to a single column

```python
print(dataf["Height"].mean())
print(dataf["Height"].sum())
print(dataf["Height"].min())
print(dataf["Height"].max())
print(dataf["Height"].count())
```

Now, we can group the data in DataFrame by some specific column using the **groupby()** function.   
We can then use the **aggregate function** on this group to get more specific information about the grouped column type data. 

```python
group = dataf.groupby("Type1")
print(group["Height"].mean())
```

```python
print(group["Height"].sum())
```

This is how we can go on with aggregate functions for groupby() function.

Let's now roll with Data cleaning using Pandas.  
Data cleaning is removing/fixing any irrelevant, incomplete, incorrect data. ~75% of Data cleaning is done in pandas

```python
# 1. Drop irrelevant columns
# We can use 'drop(columns=[])' to drop the columns. just need to enter the columns in the list.
dataf = dataf.drop(columns=["Legendary"]) # This functions assigns a new DataFrame, so we need to reassign it there.
dataf 
```

```python
# 2. Handling Missing data
# In the data, There were some pokemons which didn't have any Type2. If we have to remove those pokemons:
datafq = dataf.dropna(subset=["Type2"])
datafq
```

```python
# If we don't want to remove the missing values but instead fill them:
dataf = dataf.fillna({"Type2" : "None"}) # We are replacing NaN values with 'None' in Type2 column
print(dataf.to_string())
```

```python
# 3. Fix inconsistent values.
# In order to fix inconsistent values, we can use .replace({"x" : "X"}) to replace 'x' to "X" for a particular column.
dataf["Type1"] = dataf["Type1"].replace({"Grass" : "GRASS","Fire":"FIRE"})

print(dataf.to_string())
```

```python
# 4. Standardize text
dataf["Name"] = dataf["Name"].str.lower()

print(dataf.to_string())
```

```python
# 5. Fix Data types
dt = pd.read_csv("pokemon.csv")

dt["Legendary"] = dt["Legendary"].astype(bool)
dt
```

```python
# 6. Remove Duplicate Values
dt = dt.drop_duplicates()
dt
```
